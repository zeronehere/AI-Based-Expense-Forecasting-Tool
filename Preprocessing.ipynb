```python



from google.colab import drive
drive.mount('/content/drive')


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

file_path = '/content/drive/MyDrive/Datasets/main_dataset.csv'
df = pd.read_csv(file_path)

print("✅ Data loaded successfully!")
print(f"Initial shape: {df.shape}")
display(df.head())

#  Basic Info and Missing Values

print("\n📊 Basic Information:")
print(df.info())

print("\n🔍 Missing Values:")
print(df.isnull().sum())

# Fill missing numeric values with median

df = df.fillna(df.median(numeric_only=True))

# Fill missing categorical values with mode

for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\n✅ Missing values handled.")

#  Remove Duplicates

before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print(f"🧾 Removed {before - after} duplicate rows.")

# Fix Data Types

for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = df[col].astype(float)
        except:
            pass
print("✅ Data types corrected where possible.")

# Outlier Treatment

def cap_outliers(series):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    return np.clip(series, lower, upper)

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    df[col] = cap_outliers(df[col])
print("✅ Outliers capped (IQR method applied).")

#  Encode Categorical Columns

label_cols = ['Occupation', 'City_Tier']
le = LabelEncoder()
for col in label_cols:
    if col in df.columns:
        df[col] = le.fit_transform(df[col])
print("✅ Categorical columns encoded.")

# Feature Engineering

expense_cols = [
    'Rent','Loan_Repayment','Insurance','Groceries','Transport',
    'Eating_Out','Entertainment','Utilities','Healthcare',
    'Education','Miscellaneous'
]

df['Total_Expenses'] = df[expense_cols].sum(axis=1)
df['Actual_Savings'] = df['Income'] - df['Total_Expenses']
df['Savings_Rate'] = (df['Actual_Savings'] / df['Income']) * 100

df['Essential_Expenses'] = df[['Rent','Groceries','Utilities','Transport','Insurance']].sum(axis=1)
df['Lifestyle_Expenses'] = df[['Eating_Out','Entertainment','Miscellaneous']].sum(axis=1)
df['LongTerm_Expenses'] = df[['Loan_Repayment','Education','Healthcare']].sum(axis=1)

print("✅ Feature engineering complete!")

#  Logical Corrections

df.loc[df['Actual_Savings'] > df['Income'], 'Actual_Savings'] = 0
df.loc[df['Actual_Savings'] < 0, 'Actual_Savings'] = 0
df['Overspending'] = np.where(df['Total_Expenses'] > df['Income'], 1, 0)

print("✅ Logical data corrections done.")

# STEP 10: Correlation Check

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()


#  Scaling (Normalize numeric data for ML)

scaler = MinMaxScaler()
scaled_cols = df.select_dtypes(include=['float64','int64']).columns
df_scaled = df.copy()
df_scaled[scaled_cols] = scaler.fit_transform(df[scaled_cols])

print("✅ Data normalization (MinMax scaling) applied.")


# Final Cleaned Dataset

print(f"\nFinal shape after preprocessing: {df_scaled.shape}")
print("✅ Data preprocessing complete and ready for ML!")

# Save processed dataset
df_scaled.to_csv("/content/drive/MyDrive/cleaned_expense_data.csv", index=False)
print("\n💾 Cleaned dataset saved as 'cleaned_expense_data.csv'.")

# Quick Summary Statistics

print("\n📈 Summary Statistics:")
display(df.describe().T)
```
